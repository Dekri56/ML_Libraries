```python
import seaborn as sns
import pandas as pd
```

```python
sns.get_dataset_names()
```

```python
mpg = pd.read_csv("mpg.csv")
```

```python
mpg.head()
```

```python
sns.scatterplot(x="mpg", y="cylinders",data=mpg, hue="model_year")
```

```python
sns.histplot(mpg['horsepower'], kde=True, bins=50)
```

```python
mpg.info()
```

```python
sns.pairplot(mpg.select_dtypes(['number']), hue='model_year')
```

```python
mpg.corr()
```

## The heatmap uses color to represent the strength of the correlations, with darker colors representing stronger correlations.
## This code will create a heatmap with a range of colors from -1 to 1, where -1 represents a perfect negative correlation, 0 represents no correlation, and 1 represents a perfect positive correlation.

```python
sns.heatmap(mpg.corr(), annot=True, cmap= "YlGnBu")
```
